```python
#initializing
pygame.init()

#create window
window = pygame.display.set_mode( (300, 300) )

#FPS
FPS = 30
FPS_CLOCK = pygame.time.Clock()

#the entire game is within a while loop. 
#keyboard inputs
while 1:
    for event in pygame.event.get(): #our event while loop
        if event.type == pygame.QUIT: 
            print("Shutting down pygame")
            pygame.quit()
            exit()
        if event.type == pygame.KEYUP:
            if event.key == K_a:
                print("Key A")
        if event.type == pygame.KEYDOWN:
            if event.key == K_UP:
                print("Arrow UP has been pressed")
            if event.key == K_DOWN:
                print("Arrow Down has been pressed")
            if event.key == K_LEFT:
                print("Arrow Left has been pressed")
            if event.key == K_1:
                print("Number 1 has been pressed") 
    pressed_keys = pygame.key.get_pressed()

    if pressed_keys[K_a] and pressed_keys[K_b]:
        print("Key A and Key B has been pressed")

        pygame.display.update()
        FPS_CLOCK.tick(60)
```

```python
#commands that tells what to do when each button is pressed
def rightanswer():
    print("Button has been clicked")


def wronganswer1():
    print("wrong answer")

def wronganswer2():
    print("wrong answer")

def wronganswer3():
    print("wrong answer")
    
```

```python
#defining the button class
class Button:
    def __init__(self, x, y, width, height, text, command): #initalizes the original state of the class; sets background, font, and text color to black
        self.command = command
        
        self.bg = pygame.Rect(x, y, width, height)

        self.font = pygame.font.SysFont("Verdana", 20) 
        
        self.bttnText = self.font.render(text, True, (0,0,0)) 

    def update(self, mousePos): #detects when the mouse and button "collide", so the button's command can be executed
        if self.bg.collidepoint(mousePos): 
            self.command()
            
            
        
    def render(self, display): #gets the center of the button, sets button to white color, and displays the text in center
        self.posRect = self.bttnText.get_rect(center = (self.bg.x + self.bg.width/2, self.bg.y + self.bg.height/2)) #returns the center of the button
        pygame.draw.rect(display, (255,255,255), self.bg) 
        display.blit(self.bttnText, self.posRect) 

    def change_text(newtext): # allows me to change the text in a button as needed
        self.text= "newtext"
```

```python
#making a specific button
pygame.init() 
display = pygame.display.set_mode( (400, 300) )

right_answer = Button(x = 200, y= 100, width = 80, height= 30, text= "Button", command = rightanswer)

wrong_answer1 = Button(x = 100, y= 100, width = 80, height= 30, text= "Button2", command = wronganswer1)


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT: #necessary to actually close the game
            print("Shutting down pygame")
            pygame.quit()
            exit()  
        if event.type == pygame.MOUSEBUTTONDOWN: #defines what should happen if the button is pressed, using the update function of the button class
            right_answer.update(pygame.mouse.get_pos())
            wrong_answer1.update(pygame.mouse.get_pos())
            
            
    right_answer.render(display) #displays buttons
    wrong_answer1.render(display)
    
    pygame.display.update() #updates game display based on events occuring
```
