```python
#DEFINING A BUTTON class. can be used for all the buttons in the game!
class Button:
    def __init__(self, x, y, width, height, text, command): #initalizes the original state of the class; sets background, font, and text color to black
        self.command = command
        
        self.bg = pygame.Rect(x, y, width, height)

        self.font = pygame.font.SysFont("menlo", 17) 

        self.text = text
        self.text_color = ((0,0,0))

    def update(self, mousePos): #detects when the mouse and button "collide", so the button's command can be executed
        if self.bg.collidepoint(mousePos): 
            self.command()

    def wrap_text(self,text): #wraps text within the width of the button
        words = text.split(" ")
        lines = [] #our list of lines
        current_line = "" 
            
        # This will build lines with the words of the text, but if the current line exceeds the button width, it is ended there and added to our list of lines
        for word in words:
            test_line = current_line + word + " " 
            if self.font.size(test_line)[0] <= self.bg.width:
                current_line = test_line
            else: 
                lines.append(current_line)
                current_line = word + " "
            
        lines.append(current_line) # adding the final line to the line list, even if it hasn't exceeded button width
        return lines

        
    def render(self, display): #renders text, sets background to white color
        pygame.draw.rect(display, (255,255,255), self.bg) 

        lines = self.wrap_text(self.text) #using wrap text function 
        
        x_offset = self.bg.x +5
        y_offset= self.bg.y +5 # some padding for the text

        for line in lines:
            rendered = self.font.render(line, True, self.text_color) 
            display.blit(rendered, (x_offset, y_offset)) #displaying the text with the padding
            y_offset += rendered.get_height() + 2
        
    def change_text(self,newtext): # allows me to change the text of a button as needed
         self.text= newtext
         self.bttnText = self.font.render(newtext, True, (0,0,0))
        
    def change_position(self,new_x, new_y): #to randomize button positions. don't want the correct answer to be in the same place everytime!
            self.bg.x = new_x
            self.bg.y = new_y


        
```

```python
#BUTTON FUNCTIONS, each of which is a command that tells the game what to do when each button is pressed

#Buttons for answer choices:
    
def correctanswer():
    global subgame_state #choosing the correct answer (or the wrong answer) will change the global question number (aka the subgame state)
    global player_score
    subgame_state += 1 
    player_score += 1

def wronganswer():
    global subgame_state
    global player_score
    subgame_state += 1

#Buttons for non trivia choices:

def savetheship():
    global subgame_state
    global player_score
    global choicesmade #variable that keeps track of how many non-trivia choices have been made
    player_score -= 2
    subgame_state += 1
    choicedictionary["choice1"] +=1
    choicesmade +=1

def dontsavetheship():
    global subgame_state
    global choicesmade
    subgame_state += 1
    choicedictionary["choice2"] +=1
    choicesmade += 1

def idontknow():
    global subgame_state
    global choicesmade
    subgame_state +=1
    choicesmade += 1

def trytosavemoreships():
    global subgame_state
    global choicesmade
    global player_score
    subgame_state +=1
    choicesmade += 1
    player_score -=4
    choicedictionary["choice1"] +=1

def fixoursystems():
    global subgame_state
    global choicesmade
    subgame_state +=1
    choicesmade += 1
    choicedictionary["choice2"] +=1
    
#Transition Buttons:

def EarthLaunch():
    display.fill((0,0,0)) #clears the screen
    pygame.display.flip()
    global game_state
    game_state = "MarsIntro" #changes the game state variable

def MarsQuestions():
    display.fill((0,0,0))
    pygame.display.flip()
    global game_state 
    game_state = "MarsQuestions" 
    
def JupiterQuestions():
    display.fill((0,0,0))
    pygame.display.flip()
    global game_state
    game_state = "JupiterQuestions" 

def SaturnQuestions():
    display.fill((0,0,0))
    pygame.display.flip()
    global game_state
    game_state = "SaturnQuestions"

def UranusQuestions():
    display.fill((0,0,0))
    pygame.display.flip()
    global game_state
    game_state = "UranusQuestions"

def gameend():
    display.fill((0,0,0))
    pygame.display.flip()
    global game_state
    game_state = "gameend"
```
